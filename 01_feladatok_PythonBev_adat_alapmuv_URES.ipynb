## 4.&nbsp;Feladat: Változók, műveletek és azok sorrendje

Oldjuk meg az $\ x^2 + x-6 =0\ $ másodfokú egyenletet! 
- Ehhez először vegyük fel az egyenlet együtthatóit változóként! ```a = 1; b = 1; c = -6```
- Ellenőrzésként számítsuk ki az egyenlet diszkriminánsát és tároljuk el változóban!  $$D = \sqrt{b^2-4\cdot a\cdot c}$$
- Adjuk meg a két megoldást a megoldóképletet használva egy-egy változóban tárolva! $$x_{1,2} = \frac{-b\pm \sqrt{b^2-4\cdot a\cdot c}}{2a} = \frac{-b\pm \sqrt{D}}{2a}$$
- Majd írjuk ki a képernyőre a megoldást!

Ügyeljünk a műveletek sorrendjére, zárójelezésre! A gyökvonás megfelel 1/2 kitevőjű hatványozásnak.

Mit kapunk eredményül a diszkriminánsra és mit ad a megoldóképletünk, ha $-6$ helyett $+6$ az utolsó tag? ($\ x^2 + x+6 =0\ $)
