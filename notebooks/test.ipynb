## Gyakorlás: Premier League tabella

A pl.txt szövegfájl a Premier League 2011-12-es szezonjának eredményeit tartalmazza. Készíts programot, amely...

- kiírja, hogy a mérkőzések hány százalékán esett gól,
- kiírja, hogy melyik mérkőzésen esett a legtöbb gól,
- bekéri a felhasználótól n értékét, majd kiírja a bajnokság állását az n. forduló után (rendezési szempontok: pontszám, gólkülönbség, több rúgott gól)!

```python
# Adatok beolvasása fájlból szótárak listájába.
games = []

with open('pl.txt') as f:
    # első 6 sor átugrása
    for i in range(6):
        f.readline()

    # további sorok feldolgozása

    for line in f:
        tok = line.split('\t')
        game = {
            'round': int(tok[0]),
            'hteam': tok[1],
            'ateam': tok[2],
            'hgoals': int(tok[3]),
            'agoals': int(tok[4])
        }
        games.append(game)
```

```python
merkozes_szam = len(games)
```

```python
# A mérkőzések hány százalékán esett gól?
gol_esett_szam = sum(g['hgoals'] + g['agoals'] > 0 for g in games)
print('A mérkőzések %.1f százalékán esett gól.'%(gol_esett_szam/merkozes_szam *100))
```

```python
# Melyik mérkőzésen esett a legtöbb gól?
```

```python
# ...ugyanez tömörebben:
max(games, key=lambda x: x['hgoals'] + x['agoals'])
```

```python
# Tabella az n. forduló után.

def update_stats(hteam, ateam, hgoals, agoals, stats):
    # pontszám frissítése
    if hgoals > agoals: stats[hteam][0] += 3
    elif hgoals == agoals: stats[hteam][0] += 1
        
    # gólkülönbség frissítés
    stats[hteam][1] += hgoals - agoals
    
    # rúgott gólok frissítése
    stats[hteam][2] += hgoals
    
# n bekérése
n = int(input('n: '))

# csapatstatisztikák (pontszám, gólkülönbség, rúgott gólok) inicializálása

stats = {g['hteam']: [0, 0, 0] for g in games}

# csapatstatisztikák kiszámítása
for g in games:
    if g['round'] <= n:
        update_stats(g['hteam'], g['ateam'], g['hgoals'], g['agoals'], stats)
        update_stats(g['ateam'], g['hteam'], g['agoals'], g['hgoals'], stats)
```

```python
# rendezés
standings = sorted(stats.items(), key=lambda p: tuple(p[1]), reverse=True)
```

```python
# kiírás
for s in standings:
    team = s[0]
    pts = s[1][0]
    gd =  s[1][1]
    gs = s[1][2]    
    print(f'{team:25}{gd:5}{gs:5}{pts:5}')
```

```python
# ...tömörebben:
for team, (pts, gd, gs) in standings:
    print(f'{team:25}{gd:5}{gs:5}{pts:5}')
```
